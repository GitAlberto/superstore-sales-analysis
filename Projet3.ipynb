Observation de valeurs abérrantes des colonnes numériques du datatset (s'il y en a)

```python
# Creation de la fonction pour observer les valeurs aberantes
def obs_outliers(data,column):
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)

  # Difference interquartile
  IQR = Q3 - Q1

  # Definition des bornes
  borne_inf = Q1 - 1.5 * IQR
  borne_sup = Q3 + 1.5 * IQR

  # Recuperation de ces valeurs dans une liste
  liste_outliers = data[(data[column] < borne_inf) | (data[column] > borne_sup)][column].tolist()
  return liste_outliers
```
